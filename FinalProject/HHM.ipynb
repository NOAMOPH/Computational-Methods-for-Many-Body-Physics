# DMRG Holstein model

The Hamiltonian is
$$
H = 
-t\sum_i c^\dagger_i c_{i+1}
+ \hbar \omega_0 \sum_i b^\dagger_i b_i
- g\omega_0 \sum_i (b^\dagger_i + b_i) n_i
+
h.c.
$$


```julia
function HHM_Hamiltonian(sites,U,omega0,g)
    """
    t = 1
    """
    N = div(length(sites),2)
    ampo = OpSum()

    for i = 1:N-1
        ampo += "Adag",2*i-1,"A",2*i+1
        ampo += "Adag",2*i+1,"A",2*i-1
        # ampo += "Cdag",2*i-1,"C",2*i+1
        # ampo += "Cdag",2*i+1,"C",2*i-1
    end
    for i = 1:N
        ampo += omega0, "N",2*i
        ampo += -g*omega0, "Adag", 2*i, "N", 2*i-1
        ampo += -g*omega0, "A", 2*i, "N", 2*i-1
    end
    return MPO(ampo,sites)
end
```

```julia
Nsites = 5
# sites = siteinds(n->isodd(n) ? "Fermion" : "Boson",2*N)
# How to hold properly QN's?
sites = Vector{Index}(undef,2*Nsites)
for n=1:Nsites
    sites[2*n-1] = siteind("Boson"; addtags="n=$n",conserve_qns=true)
    # sites[2*n] = siteind("Boson",dim=3; addtags="n=$n",conserve_qns=true)
    sites[2*n] = Index([QN() => 3],"Boson,Sie,n=$n")
end
@show sites;
H = HHM_Hamiltonian(sites,1,1,1);

# Index([QN("Nc",0,-1)=>1, QN("Nc",1,-1)=>2, QN("Nc",2,-1)=>1];tags="Electron,Site,n=$n")
```

```julia
N = 10
chi = 10
s1 = siteinds("Fermion", N ; conserve_qns=true, qnname_nf="Ne")
s2 = siteinds("Boson", N ; conserve_qns=false)
psi1 = randomMPS(s1, j -> isodd(j) ? "1" : "0"; linkdims=chi)
psi2 = randomMPS(s2, j -> "0", linkdims=chi)
@show linkdims(psi1)
@show linkdims(psi2)

psi_paired = MPS([psi1[j] * psi2[j] for j in 1:N])
psi_paired = orthogonalize(psi_paired, 1)
psi = MPS(mapreduce(vcat, 1:N) do j
  psi_j, psi_jp1 = factorize(psi_paired[j], [linkinds(psi_paired, j - 1); s1[j]])
  return [psi_j, psi_jp1]
end)
@show linkdims(psi)
```

```julia
sites = siteinds("Fermion",10; conserve_qns=true)

H = OpSum()
for n in 1:9
    H += "c",n ,"c",n+1
    H += "Cdag",n, "Cdag",n+1
end
Ham = MPO(H, sites)
state0 = [isodd(n) ? "0" : "0" for n=1:10]
psi0 = randomMPS(sites,state0)

E,psi=dmrg(Ham, psi0, Sweeps(10, maxdim=10))
```
